```python
import pandas as pd

# Loading the car data from the file

file_path = 'car data.csv'
car_data = pd.read_csv(file_path)

# Displaying the  first few rows for understanding its structure
car_data.head()
```

```python
# Check for missing values in the dataset....

missing_values = car_data.isnull().sum()

# Display the missing values.........
missing_values
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the styles for  the plots......
sns.set(style="whitegrid")

# Listing the  numerical features
numerical_features = ['Year', 'Present_Price', 'Driven_kms', 'Selling_Price']

# Plotting the distribution of each numerical feature...

plt.figure(figsize=(12, 8))
for i, feature in enumerate(numerical_features, 1):
    plt.subplot(2, 2, i)
    sns.histplot(car_data[feature], kde=True)
    plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()

# Box plots are used to  identify the  outliers........

plt.figure(figsize=(12, 8))
for i, feature in enumerate(numerical_features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=car_data[feature])
    plt.title(f'Box Plot of {feature}')
plt.tight_layout()
plt.show()

# Scatter plots used to  explore relationships with Selling_Price............

plt.figure(figsize=(12, 8))
for i, feature in enumerate(numerical_features[1:], 1):  # Exclude Selling_Price itself
    plt.subplot(2, 2, i)
    sns.scatterplot(x=car_data[feature], y=car_data['Selling_Price'])
    plt.title(f'Selling_Price vs {feature}')
plt.tight_layout()
plt.show()

# Bar charts are used for features of the cars.........
categorical_features = ['Fuel_Type', 'Selling_type', 'Transmission']

plt.figure(figsize=(12, 4))
for i, feature in enumerate(categorical_features, 1):
    plt.subplot(1, 3, i)
    sns.countplot(x=car_data[feature])
    plt.title(f'Count of {feature}')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
