```python
import pandas as pd

# Loading the dataset....!

file_path = 'spam.csv'
data = pd.read_csv(file_path, encoding='latin-1')


print('\n***    DATA PREPOCESSING    *** \n')



# Printing the columns to see what data is available...

print("Columns in the dataset:")
print(data.columns)

# Selecting the empty columns to remove from the dataset using Drop().....

columns_to_drop = ['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4']

data = data.drop(columns=[col for col in columns_to_drop if col in data.columns])

# Renaming the columns v1 and v2 ad label and text for better understanding....
print('\n Renaming the columns v1 and v2 ad label and text for better understanding')
if 'v1' in data.columns and 'v2' in data.columns:
    data = data.rename(columns={'v1': 'label', 'v2': 'text'})

# Checking  for the missing values in the dataset...... 
missing_values = data.isnull().sum()

# Print the missing values information........
print("\nMissing Values:")
print(missing_values.to_string())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report


# Split the dataset into training set and testing set.......

X_train, X_test, y_train, y_test = train_test_split(data['text'], data['label'], test_size=0.2, random_state=42)

# Initialize TF-IDF Vectorizer.....
vectorizer = TfidfVectorizer(stop_words='english')

# Fit and transform the training data, transform the testing data....
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Initialize the Multinomial Naive Bayes model........
model = MultinomialNB()

# Training the model......
model.fit(X_train_tfidf, y_train)

# Predict on the testing set........
y_pred = model.predict(X_test_tfidf)

# Evaluating the model......
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

# Print the accuracy and classification report
print('\n***    REPORT    *** \n')
print(f"Accuracy: {accuracy:.4f}\n")
print("Classification Report:\n")
print(report)
```
